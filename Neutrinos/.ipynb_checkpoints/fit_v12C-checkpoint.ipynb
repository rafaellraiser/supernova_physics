# Polynomial fit for neutral current v+12C -> v+12C* reaction

```python
import numpy as np
from matplotlib import pyplot as plt

# Dados tirados da tese do Kemp
Ev=np.array([16,18,20,22,24,26,28,30,35,40,45,50,60,70,80,90,100])
cs_nc_nu=1e-42*np.array([0.01,0.106,0.302,0.599,0.994,1.49,2.07,2.74,4.78,7.26,10.1,13.1,19.5,25.4,30.2,33.7,35.8])
cs_nc_nubar=1e-42*np.array([0.0095,0.099,0.279,0.547,0.896,1.32,1.82,2.38,4.03,5.95,8.03,10.2,14.4,17.9,20.7,22.5,23.6])
cs_cc_nue=1e-42*np.array([0,0.036,0.287,0.772,1.49,2.44,3.62,5.03,9.47,15.1,21.8,29.2,45.2,60.8,74.2,84.2,90.6])
cs_cc_nuebar=1e-42*np.array([0.086,0.327,0.711,1.23,1.87,2.62,3.48,4.42,7.10,10.1,13.2,16.4,22.2,27.0,30.5,32.8,34.2])
```

```python
pol_nc_nu=np.poly1d(np.polyfit(Ev,cs_nc_nu,5))
pol_nc_nubar=np.poly1d(np.polyfit(Ev,cs_nc_nubar,5))

pol_cc_nue=np.poly1d(np.polyfit(Ev,cs_cc_nue,5))
pol_cc_nuebar=np.poly1d(np.polyfit(Ev,cs_cc_nuebar,5))

x=np.linspace(0,100,1000)
plt.plot(Ev,cs_nc_nu,'.',x,pol_nc_nu(x))
plt.plot(Ev,cs_nc_nubar,'.',x,pol_nc_nubar(x))
plt.plot(Ev,cs_cc_nue,'.',x,pol_cc_nue(x))
plt.plot(Ev,cs_cc_nuebar,'.',x,pol_cc_nuebar(x))
plt.show()
```

```python
error_nc_nu=np.zeros(len(Ev))
error_nc_nubar=np.zeros(len(Ev))
error_cc_nue=np.zeros(len(Ev))
error_cc_nuebar=np.zeros(len(Ev))

for n in range(len(Ev)):
    error_nc_nu[n]=np.abs( pol_nc_nu(Ev[n])-cs_nc_nu[n] )/cs_nc_nu[n]
    error_nc_nubar[n]=np.abs( pol_nc_nubar(Ev[n])-cs_nc_nubar[n] )/cs_nc_nubar[n]
    error_cc_nue[n]=np.abs( pol_cc_nue(Ev[n])-cs_cc_nue[n] )/pol_cc_nue(Ev[n])
    error_cc_nuebar[n]=np.abs( pol_cc_nuebar(Ev[n])-cs_cc_nuebar[n] )/cs_cc_nuebar[n]

plt.plot(Ev,error_nc_nu,'.',Ev,error_nc_nubar,'.',Ev,error_cc_nue,'.',Ev,error_cc_nuebar,'.')
plt.legend(['nc_nu','nc_nubar','cc_nue','cc_nuebar'])
```
